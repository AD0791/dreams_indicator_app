```python
%%capture
datim = AgywPrev()
```

```python
datim.data_dreams_valid
datim.total_datim_general
df = datim.data_dreams_valid
```

```python
df_sample  = df.sample(frac=0.1,random_state=1)
df_sample
```

```python
resume_sample = df_sample.pivot_table(values="id_patient",index='age_range',columns='departement',aggfunc='count',fill_value=0,margins=True,margins_name="Total")
resume_sample
```

```python
artibonites = df_sample[df_sample.departement == "Artibonite"]
nord = df_sample[df_sample.departement == "Nord"]
ouest = df_sample[df_sample.departement == "Ouest"]

artibonites.id_patient.count()
nord.id_patient.count()
ouest.id_patient.count()
```

```python
sampling = ExcelWriter("echantillon_Q1FY22_december.xlsx",engine="xlsxwriter")
#sampling = ExcelWriter("echantillon_Q1FY22.xlsx",engine="xlsxwriter")
resume_sample.to_excel(sampling,sheet_name="Tableau Resume", index=True)
artibonites.to_excel(sampling,sheet_name="Artibonite", index=False, na_rep="")
nord.to_excel(sampling,sheet_name="Nord", index=False, na_rep="")
ouest.to_excel(sampling,sheet_name="Ouest", index=False, na_rep="")
sampling.save()
```
