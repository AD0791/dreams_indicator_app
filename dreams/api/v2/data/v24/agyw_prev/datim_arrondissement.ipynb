```python
#wb.sheetnames

def name_handler(s):
    s = sub(r"[^\w\s]", '', s)
    # Replace all runs of whitespace with a single dash
    s = sub(r"\s+", '_', s)
    #remove \
    s = sub(r"\/",'',s)
    return s

```
